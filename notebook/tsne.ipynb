```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# 1. Chargement des données [cite: 9]
df = pd.read_csv('city_lifestyle_dataset.csv')
# Note: Assurez-vous de ne garder que les colonnes numériques pour le t-SNE
df_numeric = df.select_dtypes(include=['number']).dropna()

# 2. Implémentation du t-SNE (Projection 2D) [cite: 33, 36]
tsne = TSNE(n_components=2, random_state=42)
X_embedded = tsne.fit_transform(df_numeric)

# 3. Visualisation
plt.figure(figsize=(10, 7))
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], alpha=0.7)
plt.title("Visualisation t-SNE du City Lifestyle Dataset")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.show()

# 4. Exportation des données [cite: 37, 20]
output_df = pd.DataFrame(X_embedded, columns=['tsne_1', 'tsne_2'])
output_df.to_csv('tsne_emb_2d.csv', index=False)
```

Analyse du résultat

Structure globale : On observe une séparation distincte des données en plusieurs groupes ou "clusters".

Détails : Il existe un groupe très dense et étendu sur la gauche (entre -15 et 0 sur la Dimension 1) et une structure plus allongée sur la droite.

Cela suggère que le dataset contient des catégories de villes ayant des caractéristiques de style de vie très similaires entre elles, mais nettement différentes des autres groupes.
